```python
N = 64
dim = N
sub_exp = f"N{N}_T128_noisy"
activation = "recttanh"
folder = os.path.join(exp_folder, sub_exp, activation)
trajectories_target = np.load(os.path.join('experiments', 'avi', 'all_trajs', f'N{N}_{activation}', 'trajectories_2024-05-21-19-49-33.npy'))
trajectories_target_full, trajectories_target, mean, std = normalize_scale_pair(torch.tensor(trajectories_target), training_pairs=False)
folder
```

```python
hdsn_path = os.path.join(motif_dir, 'ring', "homeo_trajectories_2024-05-21-19-49-33.pth")
homeo_params = {'homeo_type': 'node'}
with open(os.path.join(motif_dir, 'ring', 'parameters_trajectories_2024-05-21-19-49-33.pkl'), "rb") as f:
    all_parameters = pickle.load(f)
ds_params = all_parameters['ds_params']
homeo_params['layer_sizes'] = [64]
homeo_params['dim'] = N
ds_params['dim'] = N
ds_params['dt'] = 5/128
homeo = build_homeomorphism(homeo_params)
source_system_ra = build_ds_motif(**ds_params)
homeo_ds_net = load_homeo_ds_net(hdsn_path, homeo, source_system_ra)
fit_ra_points = get_homeo_invman(homeo_ds_net.homeo_network, dim=dim)
fit_ra_points = (fit_ra_points) * std.detach().numpy() + mean.detach().numpy()
trajectories_source, transformed_trajectories, error = test_single_homeo_ds_net(homeo_ds_net, trajectories_target)
```

```python
#PCA
pca = PCA(n_components=2)
pca.fit(trajectories_target[:, :, :].reshape(-1, trajectories_target.shape[2]))
traj_pca = pca.transform(trajectories_target[:, :, :].reshape(-1, trajectories_target.shape[2])).reshape(-1, trajectories_target.shape[1], 2)
plot_pca_trajectories(traj_pca, trajectories_target, fit_ra_points, save_path=os.path.join(motif_dir, 'ring'))
```
