```python
current_dir = 'C:\\Users\\abel_\\Documents\\Rotations\\CIT\cit_for_computation'
exp_extension = "\\experiments\\dynamic_poisson_clicks\\"

exp_list = glob.glob(current_dir + exp_extension + "\\*45")
exp_path = exp_list[0]
with open(exp_path + '\\training_kwargs.pickle', 'rb') as handle:
    training_kwargs = pickle.load(handle)

training_kwargs['exp_path'] = exp_path
training_kwargs['training_weights_path'] = exp_path + '/training/'
training_kwargs['figures_path'] = exp_path + '//figures/'

N_rec = training_kwargs['N_rec']
N_in =  training_kwargs['N_in']
N_out = training_kwargs['N_out']

rnn_model = RNNModel(N_in, N_out, N_rec, n_layers=1, transform_function='relu').to(device)
rnn_model = load_model(rnn_model, training_kwargs).to(device)
exp_path
```

```python
#load
with open(exp_path + "\\dataset.pickle", 'rb') as handle:
    dataset = pickle.load(handle)
train_set, test_set = dataset
test_x, test_y, output_mask, trial_params = train_set
    
#create
# c_N_batch = 10000
# pdc = DynamicPoissonClicks(c_N_batch, training_kwargs)
# test_x, test_y, output_mask, trial_params =  pdc.get_trial_batch()

inputs = torch.tensor(test_x, dtype=torch.float)
yhat, hidden = rnn_model(inputs)
yhat = yhat.cpu().detach().numpy()

trial_params = np.array(trial_params)
```

```python
chosen = np.argmax(np.mean(yhat*output_mask, axis=1), axis = 1)
truth = np.argmax(np.mean(test_y*output_mask, axis = 1), axis = 1)
response_correctness = np.equal(truth, chosen)
error_idx = np.where(response_correctness==False)[0]
correct_idx = np.where(response_correctness==True)[0]

accuracy = pd_accuracy_function(test_y, yhat, output_mask)
print(accuracy)
```

```python
correct_trial_i = 1360
trial_i = correct_idx[correct_trial_i]
trial_params_i = trial_params[trial_i]
pplt.figure(figsize=(6,4))
ax = plt.subplot(111)

times = np.linspace(0, training_kwargs['T'], yhat[trial_i, :, 0].shape[0])
switch_times = trial_params_i['switch_times']
for st in switch_times:
    plt.axvline(x=st*training_kwargs['dt']);

ax.plot(times, np.where(inputs[trial_i,:,0]>0., inputs[trial_i,:,0], np.nan), 'xb')
ax.plot(times, np.where(inputs[trial_i,:,1]>0., inputs[trial_i,:,1], np.nan), 'xy')

ax.plot(times, test_y[trial_i,:,0], 'b-', label='Target 0')
ax.plot(times, test_y[trial_i,:,1], 'y-', label='Target 1')
ax.plot(times, yhat[trial_i,:,0], 'b--', label='Network output 0')
ax.plot(times, yhat[trial_i,:,1], 'y--', label='Network output 1')

ax.set_xlabel("Time (ms)")
ax.legend(loc='upper center', bbox_to_anchor=(0.5, -0.18),
          fancybox=True, shadow=True, ncol=5)

plt.savefig(training_kwargs['figures_path'] + "\\example_correct_trial.pdf")
plt.savefig(training_kwargs['figures_path'] + "\\example_correct_trial.png")
```

```python
#psychometric curve: ideal log-odds for go right vs prob go right (FIG 1d)

# plt.savefig(training_kwargs['figures_path'] + "\\psycho_logodds_probright.pdf")
# plt.savefig(training_kwargs['figures_path'] + "\\psycho_logodds_probright.png")
```

```python
#psychometric curve: final state duration vs correctness (FIG 1e)
final_state_duration = [trial_params[i]['stim_duration']-trial_params[i]['switch_times'][-1] if len(trial_params[i]['switch_times'])>0 else trial_params[i]['stim_duration'] for i in range(len(trial_params))]
stim_duration = [trial_params[i]['stim_duration'] for i in range(len(trial_params))]
df = pandas.DataFrame({"chosen": chosen, "truth": truth,
                       "response_correctness":response_correctness,
                       "final_state_duration":final_state_duration,
                      "stim_duration": stim_duration})

bins = np.linspace(0, 200, 11)
correct_per_duration = df.groupby(pandas.cut(df['final_state_duration'], bins=bins))['response_correctness'].mean()
std = df.groupby(pandas.cut(df['final_state_duration'], bins=bins))['response_correctness'].std().values
std = np.where(std+correct_per_duration>1, 1-correct_per_duration, std)
plt.errorbar((bins[:-1]+10)*training_kwargs['dt'], correct_per_duration.values, yerr=std);
plt.xlabel("Final state duration (ms)");
plt.ylabel("Prob. correct");
plt.subplots_adjust(left=0.16);

plt.savefig(training_kwargs['figures_path'] + "\\psycho_correct_final.pdf")
plt.savefig(training_kwargs['figures_path'] + "\\psycho_correct_final.png")
```

```python
#correlation between last click and choice
last_left_clicks = []
last_right_clicks = []
for trial_i in range(test_x.shape[0]):
    try:
        last_left_clicks.append(np.where(test_x[trial_i,:,0]>0.)[0][-1])
    except:
        last_left_clicks.append(0)
    try:
        last_right_clicks.append(np.where(test_x[trial_i,:,1]>0.)[0][-1])
    except:
        last_right_clicks.append(0)
        
last_left_clicks = np.array(last_left_clicks)
last_right_clicks = np.array(last_right_clicks)
```

```python
# last_click = np.where(last_left_clicks-last_right_clicks>0, 0)
df_last = pandas.DataFrame(data={"last_left":last_left_clicks, "last_right":last_right_clicks, "last_time_diff": last_left_clicks-last_right_clicks, "chosen": 2*(chosen-.5)})
```

```python
np.where(last_left_clicks==last_right_clicks)
# conditions  = [ np.where(last_left_clicks>last_right_clicks)[0], np.where(last_left_clicks==last_right_clicks)[0], np.where(last_left_clicks<last_right_clicks)[0]]
conditions = [df_last['last_left']>df_last['last_right'], df_last['last_left']==df_last['last_right'],df_last['last_left']<df_last['last_right']]
last_click_types = ["left", "both", "right"] #left, both, right
df_last['last_type'] = np.select(conditions, last_click_types, default=np.nan)
```

```python
table = df_last.chosen.groupby([df_last.last_type, df_last.chosen]).sum().abs().unstack().fillna(0).astype(int)

table/test_x.shape[0]
```
