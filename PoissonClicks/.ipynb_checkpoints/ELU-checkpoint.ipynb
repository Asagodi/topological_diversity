```python
import scipy
from scipy import special
import numpy as np
import matplotlib.pyplot as plt

def ReLU(x):
    return np.where(x<0,0,x)

def ELU(x, alpha=1.):
    return np.where(x<0,alpha*(np.exp(x)-1),x)

def SiELU(x, alpha=1.):
    return np.where(x<0,2*sigmoid(x)*alpha*(np.exp(x)-1),x)

# Gaussian Error Linear Units https://paperswithcode.com/method/gelu#:~:text=The%20GELU%20activation%20function%20is%20x%20%CE%A6%20(%20x%20)%20%2C%20where,of%20as%20a%20smoother%20ReLU.
def sigmoid(x):
    return 1/(1+np.exp(-x))

def SiLU(x):
    return x*sigmoid(x)

def GELU(x):
    return x*1/2.*(1+special.erf(x/np.sqrt(2)))

import numpy as np
import matplotlib as mpl

import os, sys
currentdir = os.path.dirname(os.path.abspath(os.getcwd()))
parentdir = os.path.dirname(currentdir)
sys.path.insert(0, currentdir) 

# sns.set_context("notebook", font_scale=1.25, rc={"lines.linewidth": 1})
cmap = 'coolwarm'

from matplotlib.ticker import MaxNLocator
import seaborn as sns

from analysis_functions import find_analytic_fixed_points

scale = 50
vmin = -6
vmax = 1
```

```python
x = np.arange(-5,2,0.1)
plt.plot(x,ReLU(x+1), label="ReLU")
plt.plot(x,ELU(x+1,1), label="ELU")
plt.plot(x,GELU(x+1), label="GELU")
plt.plot(x,SiLU(x+1), label="SiLU")
plt.plot(x,SiELU(x+1), label="SiELU")
plt.legend()
```

```python
activation_function = ELU
levels_min = -6
levels_max = 4
```

```python
#Bounded line attractor: vector field and trajectories
sns.set_context("notebook", font_scale=1.25, rc={"lines.linewidth": 1})

W_hh = np.array([[.0,-1.0],[-1,0]])
b = np.array([1.,1.])

fig, axs = plt.subplots(1, 2, figsize=(11, 5), sharex=True, sharey=True)
axs = axs.flat
fig.suptitle('Bounded line attractor')

# Line attractor dynamics
axs[0].set_title('Vector field')

w = 2.5
axs[0].set(xlim=(-w/2, w), ylim=(-w/2, w))
Y, X = np.mgrid[-w/2:w:100j, -w/2:w:100j]
U = activation_function(W_hh[0,0]*X + W_hh[0,1]*Y + b[0]) - X
V = activation_function(W_hh[1,0]*X + W_hh[1,1]*Y + b[1]) - Y
speed = np.log(np.sqrt(U**2 + V**2))
axs[0].contourf(X, Y, speed, scale, vmin=-6, vmax=1)

Y, X = np.mgrid[-w/2:w:15j, -w/2:w:15j]
U = ELU(W_hh[0,0]*X + W_hh[0,1]*Y + b[0]) - X
V = ELU(W_hh[1,0]*X + W_hh[1,1]*Y + b[1]) - Y
U, V = U/np.sqrt(U**2 + V**2), V/np.sqrt(U**2 + V**2)
speed = np.log(np.sqrt(U**2 + V**2))
axs[0].quiver(X,Y,U,V)

axs[0].xaxis.set_major_locator(MaxNLocator(integer=True))
axs[0].yaxis.set_major_locator(MaxNLocator(integer=True))
axs[0].set_xlabel("neuron 1")
axs[0].set_ylabel("neuron 2")

Y, X = np.mgrid[-w/2:w:100j, -w/2:w:100j]
U = activation_function(W_hh[0,0]*X + W_hh[0,1]*Y + b[0]) - X
V = activation_function(W_hh[1,0]*X + W_hh[1,1]*Y + b[1]) - Y
speed = np.log(np.sqrt(U**2 + V**2))

axs[1].set_title('Trajectories')
strm = axs[1].streamplot(X, Y, U, V, density=.3, broken_streamlines=False, color='r')
contourf_ = axs[1].contourf(X, Y, speed, scale, vmin=vmin, vmax=vmax)
axs[1].set_xlabel("neuron 1")

cb_ax = fig.add_axes([0.925,.124,.02,.754])
fig.colorbar(contourf_, cax=cb_ax)
cb_ax.yaxis.set_major_locator(MaxNLocator(integer=True))
cb_ax.set_ylabel("log(speed)");

# plt.savefig(currentdir+"/PoissonClicks/figs/bounded_lineattractor.pdf", bbox_inches="tight")
# plt.savefig(currentdir+"/PoissonClicks/figs/bounded_lineattractor.png", bbox_inches="tight")
# plt.savefig(currentdir+"/Stability/figures/bounded_lineattractor.pdf", bbox_inches="tight")
# plt.savefig(currentdir+"/Stability/figures/bounded_lineattractor.png", bbox_inches="tight")
# # plt.show()
```

```python
eps11 = .0
eps12 = .1
eps21 = .1
eps22 = -0.1

def nullcline_1_pos(x1):
    return ((1-eps11)*x1+1)/(eps12-1)

def nullcline_1_neg(x1):
    return (np.log(x1+1)-eps11*x1-1)/(eps12-1)

def nullcline_1(x1):
    x12 = []
    for x1_i in x1:
        if eps11*x1_i + (eps12-1)*x1_i+1 > 0:
            x12.append([x1_i, nullcline_1_pos(x1_i)])
        else:
            x12.append([x1_i, nullcline_1_neg(x1_i)])
    return np.array(x12)

def nullcline_2_pos(x2):
    return ((1-eps22)*x2+1)/(eps21-1)

def nullcline_2_neg(x2):
    return (np.log(x2+1)-eps22*x2-1)/(eps21-1)

def nullcline_2(x2):
    x12 = []
    for x2_i in x2:
        if eps22*x2_i + (eps21-1)*x2_i+1 > 0:
            x12.append([nullcline_2_pos(x2_i),x2_i])
        else:
            x12.append([nullcline_2_neg(x2_i),x2_i])
    return np.array(x12)

x = np.arange(0,5,0.1)

xy = nullcline_1(x)
plt.plot(xy[:,0], xy[:,1], '.')

xy = nullcline_2(x)
plt.plot(xy[:,0], xy[:,1], '.')
# plt.fill_between(x, ((eps12-1)*x+1)/eps11)
```

```python
W_hh = np.array([[.0,-1.0],[-1,0]])
b = np.array([1.,1.])
eps11 = .001
eps12 = .00
eps21 = -.01
eps22 = .1
eps_mat2 = np.array([[eps11,eps12],[eps21,eps22]])
W_hh = W_hh.copy()
W_hh += eps_mat2

fig, axs = plt.subplots(1, 2, figsize=(11, 5), sharex=True, sharey=True)
axs = axs.flat
fig.suptitle('Bounded line attractor')

# Line attractor dynamics
axs[0].set_title('Vector field')
axs[0].plot(nullcline_1(x), x)
axs[0].plot(x, nullcline_2(x))
w = 2.5
axs[0].set(xlim=(-w/2, w), ylim=(-w/2, w))
Y, X = np.mgrid[-w/2:w:100j, -w/2:w:100j]
U = activation_function(W_hh[0,0]*X + W_hh[0,1]*Y + b[0]) - X
V = activation_function(W_hh[1,0]*X + W_hh[1,1]*Y + b[1]) - Y
speed = np.log(np.sqrt(U**2 + V**2))
axs[0].contourf(X, Y, speed, scale, vmin=-6, vmax=1)

Y, X = np.mgrid[-w/2:w:15j, -w/2:w:15j]
U = ELU(W_hh[0,0]*X + W_hh[0,1]*Y + b[0]) - X
V = ELU(W_hh[1,0]*X + W_hh[1,1]*Y + b[1]) - Y
U, V = U/np.sqrt(U**2 + V**2), V/np.sqrt(U**2 + V**2)
speed = np.log(np.sqrt(U**2 + V**2))
axs[0].quiver(X,Y,U,V)

axs[0].xaxis.set_major_locator(MaxNLocator(integer=True))
axs[0].yaxis.set_major_locator(MaxNLocator(integer=True))
axs[0].set_xlabel("neuron 1")
axs[0].set_ylabel("neuron 2")

Y, X = np.mgrid[-w/2:w:100j, -w/2:w:100j]
U = activation_function(W_hh[0,0]*X + W_hh[0,1]*Y + b[0]) - X
V = activation_function(W_hh[1,0]*X + W_hh[1,1]*Y + b[1]) - Y
speed = np.log(np.sqrt(U**2 + V**2))

axs[1].set_title('Trajectories')
strm = axs[1].streamplot(X, Y, U, V, density=.3, broken_streamlines=False, color='r')
contourf_ = axs[1].contourf(X, Y, speed, scale, vmin=vmin, vmax=vmax)
axs[1].set_xlabel("neuron 1")

cb_ax = fig.add_axes([0.925,.124,.02,.754])
fig.colorbar(contourf_, cax=cb_ax)
cb_ax.yaxis.set_major_locator(MaxNLocator(integer=True))
cb_ax.set_ylabel("log(speed)");

# plt.savefig(currentdir+"/PoissonClicks/figs/bounded_lineattractor.pdf", bbox_inches="tight")
# plt.savefig(currentdir+"/PoissonClicks/figs/bounded_lineattractor.png", bbox_inches="tight")
# plt.savefig(currentdir+"/Stability/figures/bounded_lineattractor.pdf", bbox_inches="tight")
# plt.savefig(currentdir+"/Stability/figures/bounded_lineattractor.png", bbox_inches="tight")
# # plt.show()
```

```python
W_hh = np.array([[.0,-1.0],[-1,0]])
b = np.array([1.,1.])

w = 2.8
Y, X = np.mgrid[-w:w:100j, -w:w:100j]
eps1 = -1
eps2 = 1
eps_mat1 = np.array([[eps1,eps2],[eps2,eps1]])/100.
W_hh_pert = W_hh
W_hh_pert += eps_mat1
b_pert = b

U = activation_function(W_hh_pert[0,0]*X + W_hh_pert[0,1]*Y + b_pert[0]) - X
V = activation_function(W_hh_pert[1,0]*X + W_hh_pert[1,1]*Y + b_pert[1]) - Y
speed = np.log(np.sqrt(U**2 + V**2))
fig, axs = plt.subplots(2, 2, figsize=(11, 10), sharex=True, sharey=True)
fig.suptitle('Types of perturbations of the bounded line attractor')
axs = axs.flat

# Line attractor dynamics
axs[0].set_title('Single fixed point') #t with full support
strm = axs[0].streamplot(X, Y, U, V, density=.3, broken_streamlines=False, color='r')
axs[0].contourf(X, Y, speed, levels=range(levels_min,levels_max))
fixed_point_list, stabilist, unstabledimensions = find_analytic_fixed_points(W_hh, b, W_ih=np.zeros((2,2)), I=[0,0], tol=10**-4)
for fxdpnt in fixed_point_list:
    axs[0].plot(fxdpnt[0], fxdpnt[1], 'o', color='pink', markersize=10)
    
axs[0].set(xlim=(-.5*w, 1.*w), ylim=(-.5*w, 1.*w))
axs[0].set_ylabel("neuron 2")
axs[0].xaxis.set_major_locator(MaxNLocator(integer=True))
axs[0].yaxis.set_major_locator(MaxNLocator(integer=True))
# axs[0].set_xlabel("neuron 1")


#
axs[1].set_title('Three fixed points')
W_hh = np.array([[.0,-1.0],[-1,0]])
b = np.array([1.,1.])
eps1 = .1
eps2 = -.02
eps_mat2 = np.array([[eps1,eps2],[eps2,eps1]])
W_hh_pert2 = W_hh
W_hh_pert2 += eps_mat2
b_pert2 = b

U2 = activation_function(W_hh_pert2[0,0]*X + W_hh_pert2[0,1]*Y + b_pert2[0]) - X
V2 = activation_function(W_hh_pert2[1,0]*X + W_hh_pert2[1,1]*Y + b_pert2[1]) - Y
speed2 = np.log(np.sqrt(U2**2 + V2**2))

strm = axs[1].streamplot(X, Y, U2, V2, density=.5, broken_streamlines=False, color='r')
contourf_ = axs[1].contourf(X, Y, speed2, levels=range(levels_min,levels_max))
#separatrix
axs[1].plot([0,w], [0,w], color='purple', linewidth=3.3)
fixed_point_list, stabilist, unstabledimensions = find_analytic_fixed_points(W_hh_pert2, b, W_ih=np.zeros((2,2)), I=[0,0], tol=10**-4)
for fxdpnt in fixed_point_list:
    axs[1].plot(fxdpnt[0], fxdpnt[1], 'o', color='pink', markersize=10)


axs[2].set_title('Two fixed points')
W_hh = np.array([[.0,-1.0],[-1,0]])
b = np.array([1.,1.])
eps11 = .00
eps12 = .00
eps21 = .0
eps22 = .1
eps_mat2 = np.array([[eps11,eps12],[eps21,eps22]])
W_hh_pert2 = W_hh.copy()
W_hh_pert2 += eps_mat2
b2 = 0.00
b_pert2 = b + [2*b2/(2+eps22),b2]

U2 = activation_function(W_hh_pert2[0,0]*X + W_hh_pert2[0,1]*Y + b_pert2[0]) - X
V2 = activation_function(W_hh_pert2[1,0]*X + W_hh_pert2[1,1]*Y + b_pert2[1]) - Y
speed2 = np.log(np.sqrt(U2**2 + V2**2))

strm = axs[2].streamplot(X, Y, U2, V2, density=.5, broken_streamlines=False, color='r')
contourf_ = axs[2].contourf(X, Y, speed2, levels=range(levels_min,levels_max))
fixed_point_list, stabilist, unstabledimensions = find_analytic_fixed_points(W_hh_pert2, b_pert2, W_ih=np.zeros((2,2)), I=[0,0], tol=10**-4)
for fxdpnt in fixed_point_list:
    print(fxdpnt)
    axs[2].plot(fxdpnt[0], fxdpnt[1], 'o', color='pink', markersize=10)
axs[2].set_xlabel("neuron 1")
axs[2].set_ylabel("neuron 2")

epsilon1=0.05
epsilon2=0.5
W_hh_pert2 = np.array([[.0+epsilon1,-1+epsilon2],[-1+epsilon1,0+epsilon2]])
axs[3].set_title('Rotated BLA')

U2 = activation_function(W_hh_pert2[0,0]*X + W_hh_pert2[0,1]*Y + b[0]) - X
V2 = activation_function(W_hh_pert2[1,0]*X + W_hh_pert2[1,1]*Y + b[1]) - Y
speed2 = np.log(np.sqrt(U2**2 + V2**2))

strm = axs[3].streamplot(X, Y, U2, V2, density=.3, broken_streamlines=False, color='r')
contourf_ = axs[3].contourf(X, Y, speed2, levels=range(levels_min,levels_max))
axs[3].set_xlabel("neuron 1")

cb_ax = fig.add_axes([0.925,.124,.02,.754])
fig.colorbar(contourf_, cax=cb_ax)
cb_ax.yaxis.set_major_locator(MaxNLocator(integer=True))
cb_ax.set_ylabel("log(speed)")

# plt.savefig(currentdir+"/Stability/figures/bounded_lineattractor_allpert.pdf", bbox_inches="tight")
# plt.savefig(currentdir+"/Stability/figures/bounded_lineattractor_allpert.png", bbox_inches="tight")
plt.show()
```
