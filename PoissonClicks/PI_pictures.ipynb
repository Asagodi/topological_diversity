```python
def ReLU(x):
    return np.where(x<0,0,x)

def ELU(x, alpha=1):
    return np.where(x<0,alpha*(np.exp(x)-1),x)

import numpy as np
import matplotlib as mpl

import os, sys
currentdir = os.path.dirname(os.path.abspath(os.getcwd()))
parentdir = os.path.dirname(currentdir)
sys.path.insert(0, currentdir) 

cmap = 'coolwarm'

from matplotlib.ticker import MaxNLocator
import seaborn as sns
sns.set_context("notebook", font_scale=1.25, rc={"lines.linewidth": 1})

from analysis_functions import find_analytic_fixed_points

scale = 50
vmin = -6
vmax = 1
```

```python
W_hh = np.array([[.1,-1.0],[-1,0]])
Wx = np.array([[-1.,0],[0,-1]])
b = np.array([1,1])
F = lambda t, x: ReLU(np.dot(W_hh, x) + b) + np.dot(Wx, x)

t_span = [0, 5]
x0 = [.4,.1]
method = 'RK45'
# t_eval = np.arange(0, 10.01, 0.01)
sol = scipy.integrate.solve_ivp(F, t_span, x0, method = method)
```
